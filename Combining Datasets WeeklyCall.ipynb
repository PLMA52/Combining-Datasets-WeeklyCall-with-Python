```python
import pandas as pd
import glob

All = pd.DataFrame()
for f in glob.glob("weekdata*.xlsx"):
    df = pd.read_excel(f)
    All = All.append(df,ignore_index=False)
    
All
```

```python
# Checking missing value
All.isnull().sum()
```

```python
# Average length of a call per day

All.groupby('DayOfWeek')['CallDuration'].mean()
```

```python
# Average length of call per agent
All.groupby('AnsweringAgent')['CallDuration'].mean()
```

```python
All.groupby('ReasonForCall')['CallDuration'].mean()
```
